```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

```

```python
df = pd.read_excel("../data/raw/data-raw.xlsx")
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
text_cols = df.select_dtypes(include='object').columns
text_cols
```

```python
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].str.replace(r'[^\x00-\x7F]+', '', regex=True)
    df[col] = df[col].str.lower()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
df = df[df['Quantity'] > 0]
```

```python
df['Price'].describe()
```

```python
df = df[df['Price'] > 0] #remove free or negative prices
df = df[df['Price'] < 10000] #remove unrealistically high prices
```

```python
df['TotalPrice'] = df['Quantity'] * df['Price']
```

```python
df = df.dropna(subset=['Customer ID'])
```

```python
df.isnull().sum()
df.describe()
df.head()
```

```python
df.to_excel("data/processed/cleaned.xlsx", index=False)
```
